```python
import numpy as np
import pandas as pd
import time
from bs4 import BeautifulSoup
import requests
```

```python
endpoints = pd.read_csv('Covers_Team_Endpoints.csv')

endpoints.head()
```

```python
year = '2018-2019'
column_names = ['Team','Date','Vs','Score','Goalie','Opp_Goalie','Moneyline','O/U']
column_names_reduced = ['Date','Vs','Score','Goalie','Opp_Goalie','Moneyline','O/U']
df = pd.DataFrame(columns=column_names,index=range(100))

for idx,tm in enumerate(endpoints['tm']):
    time.sleep(5)
    row = endpoints.loc[idx,:]
    url = 'https://www.covers.com/sport/hockey/nhl/'+year+'/'+row.endpoint+'/past-results'
    try:
        r = requests.get(url)
        data = r.text
        soup = BeautifulSoup(data,'lxml')
        table_counter=0
        for table in soup.find_all('table'):
            if table_counter==0:
                row_marker=0
                df1 = pd.DataFrame(columns=column_names_reduced,index=range(100))
                for row in table.find_all('tr')[1:]:
                    column_marker = 0
                    for col in row.find_all('td'):
                        df1.iat[row_marker,column_marker]=col.get_text(strip=True)
                        column_marker += 1
                    row_marker +=1
                print(tm+' scraped..')
                df1['Team']=[tm]*100
                df = pd.concat([df,df1])
                df.dropna(inplace=True)
                table_counter=table_counter+1
            
    except:
        print("error with "+tm)
        pass
    
df
    
```

```python
df.to_csv('test.csv',index=False)
```

```python
df1=df.copy()

df1.columns = ['Date','Moneyline','Goalie','Opp_Goalie','O/U','Score','Team','Vs']
df1['day'] = df1['Date'].map(lambda x:x.split(" ")[1])
df1['month'] = df1['Date'].map(lambda x:x.split(" ")[0])
df1['year'] = np.where((df1['month']=='Sep')|(df1['month']=='Oct')|(df1['month']=='Nov')|(df1['month']=='Dec'),2018,2019)

df1['Date']=df1['day'].astype(str)+df1['month']+df1['year'].astype(str)
df1['Date']=pd.to_datetime(df1['Date'])

df1['O/U']=df1['O/U'].map(lambda x:x.split(' ')[0])
df1
```

```python
df1.drop(columns=['day','month','year'],inplace=True)
df1
```

```python
df1.to_csv('NHL_covers_'+year+'.csv',index=False)
```
